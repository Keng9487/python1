```python
import matplotlib.pyplot as mp
values1 = [20,30,45,60]
labels = ['Nokia','Samsung','Apple','Lumia']
colors = ['yellow','green','red','blue']

figure = mp.figure()
axes = figure.add_subplot(1,1,1)
axes.pie(values1,labels=labels,colors=colors,explode=(0.3,0,0,0),shadow=True,startangle=180,autopct='%.1f%%')
mp.show()
```
